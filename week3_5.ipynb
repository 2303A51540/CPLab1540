```python
import bisect


def subset_sums(arr):
    n = len(arr)
    res = []

    for mask in range(1 << n):
        s = 0
        for i in range(n):
            if mask & (1 << i):
                s += arr[i]
        res.append(s)

    return res


def closest_subset_sum(N, S, A):

    mid = N // 2
    left = A[:mid]
    right = A[mid:]

    left_sums = subset_sums(left)
    right_sums = subset_sums(right)

    right_sums.sort()

    ans = float('inf')

    for x in left_sums:

        need = S - x
        pos = bisect.bisect_left(right_sums, need)

        if pos < len(right_sums):
            ans = min(ans, abs(x + right_sums[pos] - S))

        if pos > 0:
            ans = min(ans, abs(x + right_sums[pos - 1] - S))

    return ans


# ========== MAIN ==========

T = int(input("Enter number of test cases: "))

for _ in range(T):

    N, S = map(int, input("Enter N and S: ").split())
    A = list(map(int, input("Enter array: ").split()))

    print("Output:", closest_subset_sum(N, S, A))
```
