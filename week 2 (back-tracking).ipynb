```python
# Backtracking Assignment 3: Subset Sum Problem
# Scenario: Check if some combination of expenses matches a fixed budget

def subset_sum(arr, n, target):

    # If target becomes 0, subset is found
    if target == 0:
        return True

    # If no elements left or target becomes negative
    if n == 0 or target < 0:
        return False

    # Include the last element
    include = subset_sum(arr, n - 1, target - arr[n - 1])

    # Exclude the last element
    exclude = subset_sum(arr, n - 1, target)

    # Return True if either include or exclude works
    return include or exclude


# Input
n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter elements: ").split()))
target = int(input("Enter target sum: "))

# Function call
result = subset_sum(arr, n, target)

# Output
print(result)
```
